## Faça o mesmo usando os dados de "paises"

```r
#outros usos de lapply
x1 <- list(
  c(0.27, 0.37, 0.57, 0.91, 0.20),
  c(0.90, 0.94, 0.66, 0.63, 0.06), 
  c(0.21, 0.18, 0.69, 0.38, 0.77)
)

x1 %>% lapply(function(x, cutoff = 0.8) x[x > cutoff])
```

```r
x2 <- list(
  c(0.50, 0.72, 0.99, 0.38, 0.78), 
  c(0.93, 0.21, 0.65, 0.13, 0.27), 
  c(0.39, 0.01, 0.38, 0.87, 0.34)
)

x2 %>% lapply(function(x, cutoff = 0.6) x[x<cutoff])
```

```r
#usos de apply
```

```r
# Create the matrix
m=matrix(c(seq(from=-98,to=100,by=2)),nrow=10,ncol=10)

# Return the product of each of the rows
apply(m,1,prod)

# Return the sum of each of the columns
apply(m,2,sum)

# Return a new matrix whose entries are those of 'm' modulo 10
apply(m,c(1,2),function(x) x/10) 
```

## Regular Expressions -  Regex


* `*: matches at least 0 times.`
* ` +: matches at least 1 times.`
* `?: matches at most 1 times.`
* `{n}: matches exactly n times.`
* `{n,}: matches at least n times.`
* `{n,m}: matches between n and m times.`


```r
x = c("a", "ab", "acb", "accb", "acccb", "accccb")

str_subset(x,"a")
str_subset(x,"c")
```

```r
str_subset(x,"ac+")
```

```r
str_subset(x,"ac+b")
```

```r
str_subset(x,"ac?b")
```

```r
str_subset(x,"ac{2}b")
```

* `^` : matches the start of the string.
* `$`: matches the end of the string.
* `.`: matches any character

```r
str_subset(x,"^P")
```

```r
str_subset(x,"a$")
```

```r
str_subset(x,"[ae]$")
```

```r
str_subset(x,"^[BC]")
```

```r
str_subset(str_to_lower(x),"^[bc]")
```

```r
library(gapminder)
df = gapminder
```

```r
#Encontre os paises do gapminder que tem i ou e
str_subset(df$country,"i|e") %>% unique() %>% length

#encontre os paises do gapminder que tem i ou e e terminam com land
str_subset(df$country,"i|e*land$") %>% unique() %>% length

#encontre os paises do gapminder qeu tem i ou e, terminam com land e troque land por LAND
str_replace(str_subset(df$country,"[i|e]*land$"),'land','LAND') %>% unique
```

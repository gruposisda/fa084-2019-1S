## Adicione variáveis com `mutate()`

````r
flights %>% mutate(route = paste0(origin,'-',dest))
````

````r
flights_sml = select(flights, 
  year:day, 
  ends_with("delay"), 
  distance, 
  air_time
)
````

1. Crie a coluna `speed` (lembre que air_time está em minutos)
2. Crie a coluna `gain`, que resulta da diferença entre `dep_delay` e `arr_delay`

````r
mutate(flights_sml,
  gain = dep_delay - arr_delay,
  speed = distance / air_time * 60
)  %>% head()
````

````r
mutate(flights_sml,
  gain = dep_delay - arr_delay,
  hours = air_time / 60,
  gain_per_hour = gain / hours
) %>% head()
````

## Agregue informações com `summarise()` e `group_by()`

````r
not_cancelled <- flights %>% 
  filter(!is.na(dep_delay), !is.na(arr_delay))
````

````r
not_cancelled %>% 
  group_by(year, month, day) %>% 
  summarise(mean = mean(dep_delay))
````

````r
not_cancelled %>% 
  group_by(year, month, day) %>% 
  summarise(
    avg_delay1 = mean(arr_delay),
    avg_delay2 = mean(arr_delay[arr_delay > 0]) # the average positive delay
  )
````

````r
#sd to some destinations is more variable
not_cancelled %>% 
  group_by(dest) %>% 
  summarise(distance_sd = sd(distance)) %>% 
  arrange(desc(distance_sd))
````

````r
# When do the first and last flights leave each day?
not_cancelled %>% 
  group_by(year, month, day) %>% 
  summarise(
    first = min(dep_time),
    last = max(dep_time)
  )
````
